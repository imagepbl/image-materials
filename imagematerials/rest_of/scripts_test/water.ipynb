```python
from imagematerials.read_mym import read_mym_df
import matplotlib.pyplot as plt
```

```python
#%% read in water consumption data

# water conusmption 
# 1: electricity, 2: industrial, 3: municipality, 4: total
water_consumption = read_mym_df('../../../data/raw/SSP2/EnergyFlows/ALLWCS.out').set_index(["time", "DIM_1"])
water_withdrawel = read_mym_df('../../../data/raw/SSP2/EnergyFlows/ALLWWD.out').set_index(["time", "DIM_1"])
```

```python
electricity = water_consumption.xs(key=1, level='DIM_1')
industrial = water_consumption.xs(key=2, level='DIM_1')
municipality = water_consumption.xs(key=3, level='DIM_1')
water_cons_total = water_consumption.xs(key=4, level='DIM_1')


water_wd_total = water_withdrawel.xs(key=4, level='DIM_1')


```

```python
fig, ax = plt.subplots()             # Create a figure containing a single Axes.


ax.plot(electricity[28], label = 'electricity')
ax.plot(industrial[28], label = 'industrial')
ax.plot(municipality[28], label = 'municipality')
ax.plot(water_cons_total[28], label = 'total consumption')

ax.plot(water_wd_total[28], label = 'water withdrawl total')

ax.set_ylabel("water consumption")
ax.set_xlabel("years")

ax.legend(loc='upper left')
```
