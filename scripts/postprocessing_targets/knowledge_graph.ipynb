```python
from imagematerials.concepts import create_vehicle_graph
from imagematerials.util import import_from_netcdf
```

```python
vehicle_pre = import_from_netcdf("../../../examples/prep_vema2.nc")
```

```python
knowledge_graph = create_vehicle_graph()
knowledge_graph["Cars"].synonyms = ["car"]
knowledge_graph["Bikes"].synonyms = ["bicycles"]
knowledge_graph["Airplanes"].synonyms = ["airplane"]
knowledge_graph["Trucks"].synonyms = ["truck"]
knowledge_graph["Ships"].synonyms = ["ships"]
knowledge_graph["Trains"].synonyms = ["rail"]
knowledge_graph["Buses"].synonyms = ["bus"]
```

```python
all_output = ["car", "bicycles", "airplane", "truck", "ships", "rail", "bus"]
# WEIGHTS CALCULATION
vehicle_weights = vehicle_pre["vehicle_weights"]

# Convert the region names in xr_image_output to match the target regions
xr_type = vehicle_weights.coords["Type"].values  # Extract current region names

vehicle_type_mapping_filtered = knowledge_graph.find_relations_inverse(xr_type, all_output)
vehicle_type_mapping_filtered
```
